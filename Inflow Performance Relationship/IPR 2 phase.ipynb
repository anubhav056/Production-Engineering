# Inflow Performance Relationship for two phase

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('IPR 2.xlsx',) #reading data from excel sheet
print(df)
```

```python
#storing the data
Pr=df['Reservoir Pressure'] #static reservoir pressure (psi)
Phi=df['Porosity'] #porosity of reservoir rock(decimals)
K=df['Permeability'] #permeability of reservoir rock (md)
Bo=df['Oil formation volume factor'] #oil FVF (RB/STB)
C=df['Viscosity'] #viscosity of oil (cp)
Rw=df['Radius of wellbore'] #wellbore radius (ft)
Re=df['Radius of reservoir boundary'] #reservoir boundary radius (ft)
s=df['Skin']
h=df['Height'] #(ft)
Pwf=df['Flowing bottomhole pressure'] #flowing bottomhole pressure (psi)
```

### Productivity Index


```python
R=np.log(Re/Rw)
J=K*h/((141.2*C*Bo)*(R-3/4+s)) #STB/Day-psi
print(J)
```

### Maximum flow rate
Qomax=(J*Pr)/1.8 (STB/Day)

```python
Qo=J*Pr/1.8
print(Qo)
```

### Flow rate
Q=Qo*(1-0.2*(Pwf/Pr)-0.8*((Pwf/Pr)**2)) (STB/Day)

```python
Q=Qo*(1-0.2*(Pwf/Pr)-0.8*((Pwf/Pr)**2))
print(Q)
```

### Plottting IPR curve

```python
plt.plot(Q,Pwf,color='blue',marker='o')
plt.xlabel('Q')
plt.ylabel('Pwf')
plt.title('For 2-phase')
plt.suptitle('Inflow Performance Relationship Curve')
plt.show()
```
